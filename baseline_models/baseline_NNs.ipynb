```
# Getting the MNIST dataset
# have the one_hot argument to be true since it helps
# avoid imposing unneeded structure in the output space
# such as fictitious proximity between certain digits
# and or iterpolation between the classes

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('../data/', one_hot=True)
```

```
# assign the data into feature and labels
x_train = mnist.train.images
y_train = mnist.train.labels

x_test = mnist.test.images
y_test = mnist.test.labels
```

```
# helper functions which help create weight and bias variables at each layer
def weight_variable(shape):
  w_init = tf.truncated_normal(shape, stddev=0.02)
  return tf.get_variable('w', initializer=w_init)

def bias_variable(shape):
  b_init = tf.constant(0.0, shape=shape)
  return tf.get_variable('b', initializer=b_init)
```

```
# define the variable scopes for each layer in the fully connected net

with tf.variable_scope('input'):
  x = tf.placeholder(tf.float32, [None, 784])
  y = tf.placeholder(tf.float32, [None, 10])

with tf.variable_scope('fc1'):
  W_fc1 = weight_variable([784, 1200])
  b_fc1 = bias_variable([1200])
  h_fc1 = tf.nn.relu(tf.matmul(x, W_fc1) + b_fc1)

with tf.variable_scope('fc2'):
  W_fc2 = weight_variable([1200, 1200])
  b_fc2 = bias_variable([1200])
  h_fc2 = tf.nn.relu(tf.matmul(h_fc1, W_fc2) + b_fc2)

with tf.variable_scope('fc3'):
  W_fc3 = weight_variable([1200, 1200])
  b_fc3 = bias_variable([1200])
  h_fc3 = tf.nn.relu(tf.matmul(h_fc2, W_fc3) + b_fc3)

with tf.variable_scope('softmax'):
  W_fc4 = weight_variable([1200,10])
  b_fc4 = bias_variable([10])
  scores = tf.nn.softmax(tf.matmul(h_fc3, W_fc4) + b_fc4)
```

```
# define the variable scope for the MLE estimation
with tf.variable_scope('optimization'):
  
  cross_entropy = tf.reduce_mean(-tf.reduce_sum(y * tf.log(scores), reduction_indices=[1] ))
  
  # define training step and accuracy
  train_step = tf.train.AdamOptimizer(learning_rate=0.00001).minimize(cross_entropy)
  correct = tf.equal(tf.argmax(scores, 1), tf.argmax(y, 1))
  accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

```
# start the session after setting the initializer
init = tf.global_variables_initializer()
sess = tf.InteractiveSession()
sess.run(init)
```

```
# define the epochs and batch_size
num_epochs = 1000
batch_size = 32
```

```
# train the network 
for _ in range(num_epochs):
  permutation = np.random.permutation(x_train.shape[0])
  for i in range(0, x_train.shape[0], batch_size):
    index = permutation[i: i + batch_size]
    x_batch = x_train[index]
    y_batch = y_train[index]
    
    sess.run(train_step, feed_dict={x: x_batch, y: y_batch})
    
    loss, acc = sess.run([cross_entropy, accuracy],
                       feed_dict={x: x_train, y: y_train})
    
    print('epoch {:2d} step {:4d} loss={:.3f} accuracy={:.2f}%'
          .format(_, int(i/batch_size)+1, loss, 100*acc))
```
